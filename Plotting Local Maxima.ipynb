```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nSamples = 50
```

```python
y = np.random.randint(0,10,size=(nSamples))
x = np.arange(0,nSamples,1)
print(y.shape)
print(x.shape)
```

```python
y_peaks = []
x_peaks = []
for n in range(len(y)):
    if y[n] > y[n-1] and y[n] > y[n+1]:
        y_peaks.append(y[n])
        x_peaks.append(x[n])
```

```python
print(y_peaks, x_peaks)
```

```python
fig=plt.figure(figsize=(8, 6), dpi= 80, facecolor='w', edgecolor='k')
ax=fig.add_axes([0,0,1,1])
plt.xticks(np.arange(min(x), max(x)+1, 2.0))

plt.plot(x,y)

for i in range(len(y_peaks)):
    plt.axvline(x_peaks[i], ymax=y_peaks[i]/10, linewidth=1, color='r')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('scatter plot')

plt.show()
```
